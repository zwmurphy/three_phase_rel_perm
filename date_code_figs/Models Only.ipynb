```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
plt.rcParams['font.size']=25
```

```python
# SWI (no hydrate)
nw=3.8;
ng=2.2;
Srw=0;
Srg=0;
Sw0=np.linspace(0,1,100);


for i in Sw0:
    Sg0=1-Sw0-Srw
    krw_h0=((Sw0-Srw)/(1-Srw))**nw
    krg_h0=((Sg0-Srg)/(1-Srg))**ng 
    fw_h0=krw_h0/(krw_h0+krg_h0/.1)
    fg_h0=1-fw_h0 
    krg_lw=((Sg0-Srg)/(1-Srg))**nw
```

```python
# SWI (with hydrate) SH=30
Sh=.3
nw=3.8;
ng=2.2
Srw=0;
Srg=0;

Sw_eff=np.linspace(0,1,100)
Sw30=np.linspace(0,1-Sh,100)

for i in Sw30:
    Sg_eff=1-Sw_eff
    Sg30=1-Sw30-Sh
    krw_h30=((Sw30-Srw)/(1-Srw))**nw
    krg_h30=((Sh/(1-Sg30))*((Sg30-Srg)/(1-Srg))**nw)+((1-(Sh/(1-Sg30)))*((Sg30-Srg)/(1-Srg))**ng)
    krg_30_int=((Sg30-Srg)/(1-Srg))**ng
    fw_h30=krw_h30/(krw_h30+krg_h30/.1)
    fg_h30=1-fw_h30
    
```

```python
# SWI (with hydrate) SH=60
Sh=.6
nw=3.8;
ng=2.2
Srw=0;
Srg=0;

Sw_eff=np.linspace(0,1,100)
Sw60=np.linspace(0,1-Sh,100)

for i in Sw60:
    Sg_eff=1-Sw_eff
    Sg60=1-Sw60-Sh
    krw_h60=((Sw60-Srw)/(1-Srw))**nw
    krg_h60=((Sh/(1-Sg60))*((Sg60-Srg)/(1-Srg))**nw)+((1-(Sh/(1-Sg60)))*((Sg60-Srg)/(1-Srg))**ng)
    krg_60_int=((Sg60-Srg)/(1-Srg))**ng
    fw_h60=krw_h60/(krw_h60+krg_h60/.1)
    fg_h60=1-fw_h60
    
```

```python
# SOLID MODEL  SH = 30!! 
Sh=.3
nw=4;
ng=2;
Srw=0;
Srg=0;

Sw_eff=np.linspace(0,1,100)
Sw_s30=np.linspace(0,1-Sh,100)

for i in Sw_s30:
    Sg_eff=1-Sw_eff
    Sg_s30=1-Sw_s30
    krw_s30=(((Sw_eff-Srw)/(1-Srw))**nw)*(Sw_eff/(1-Sh))
    krg_s30=(((Sg_eff-Srg)/(1-Srw))**ng)*(Sg_eff/(1-Sh))
    krw_gc30=(((Sw_eff-Srw)/(1-Srw))**nw)*(1-Sh)**2
    krg_gc30=(((Sg_eff-Srg)/(1-Srw))**ng)*(1-Sh)**2
    krw_pf30=(((Sw_eff-Srw)/(1-Srw))**nw)*(1-Sh**2+((2*((1-Sh)**2))/(np.log(Sh))))
    krg_pf30=(((Sg_eff-Srg)/(1-Srw))**ng)*(1-Sh**2+((2*((1-Sh)**2))/(np.log(Sh))))
    fw_gc30=krw_gc30/(krw_gc30+krg_gc30/.1)
    fg_gc30=1-fw_gc30
    fw_s30=krw_s30/(krw_s30+krg_s30/.1)
    fg_s30=1-fw_s30
```

```python
# SOLID MODEL  SH = 60!! 
Sh=.6
nw=4;
ng=2;
Srw=0;
Srg=0;

Sw_eff=np.linspace(0,1,100)
Sw_s60=np.linspace(0,1-Sh,100)

for i in Sw_s60:
    Sg_eff=1-Sw_eff
    Sg_s60=1-Sw_s60
    krw_s60=(((Sw_eff-Srw)/(1-Srw))**nw)*(Sw_eff/(1-Sh))
    krg_s60=(((Sg_eff-Srg)/(1-Srw))**ng)*(Sg_eff/(1-Sh))
    krw_gc60=(((Sw_eff-Srw)/(1-Srw))**nw)*(1-Sh)**2
    krg_gc60=(((Sg_eff-Srg)/(1-Srw))**ng)*(1-Sh)**2
    krw_pf60=(((Sw_eff-Srw)/(1-Srw))**nw)*(1-Sh**2+((2*((1-Sh)**2))/(np.log(Sh))))
    krg_pf60=(((Sg_eff-Srg)/(1-Srw))**ng)*(1-Sh**2+((2*((1-Sh)**2))/(np.log(Sh))))
    fw_s60=krw_s60/(krw_s60+krg_s60/.1)
    fg_s60=1-fw_s60
```
