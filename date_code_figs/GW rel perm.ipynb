# Gas Water Relative Permeability 

```python
import matplotlib.pyplot as plt
```

```python
#DATA
Sw_exp=[1,.9, .9,.8,.73,.7,.6,.9,.8,.7,.5,.25,.81,.72,.64,.59,.40,.13]
krw_exp=[1,.68,.61,.39,.27,.14,0.09,.64,.59,.24,0.032,.01,.45,.38,.34,.15,.02,0]


Sn_exp=[0,0.13,0.16,0.26,0.41,0.55,0,0.19,0.28,0.36,0.41,0.45,0.83,0.72]
Sw_from_Sn_exp=[1-Sn for Sn in Sn_exp]
krn_exp=[0,.01,0.0162,0.0683,0.1306,0.2939, 0, .01, .03, .08, .11, .19,.73,.53]
```

```python
# Brooks Corey Model
nw=3.8
nnw=2.2
Srw=0
Srn=0.0
krw_BC=[]

Sw_BC=[i*.01 for i in range(13,101)]
Sn_BC=[(1-i) for i in Sw_BC]
   
krw_BC=[((Sw-Srw)/(1-Srw))**nw for Sw in Sw_BC]   
krn_BC=[((Sn-Srn)/(1-Srn))**nnw for Sn in Sn_BC]   
```

```python
fig=plt.figure(figsize=(15,8))
plt.rcParams['font.size']=20
sharex='True'

plt1=fig.add_subplot(1,2,2)

plt1.plot(Sw_BC,krw_BC,color='blue',linestyle='solid',linewidth=3)
plt1.plot(Sn_BC,krn_BC,color='red',linestyle='solid',linewidth=3)
#plt1.plot(Sw_exp,krw_exp,'o',color='blue',markersize=10.0,label='Wetting')
#plt1.plot(Sn_exp,krn_exp,'^',color='red',markersize=10.0,label='Non-wetting')

plt.xlim([0,1])
plt.xlabel('Saturation, (S$_w$ or S$_{g}$)')
plt.xticks([0,0.5,1.0])

plt.ylim([0,1])
#plt.ylabel('relative permeability, (k$_{rw}$ or k$_{rg}$)')
plt.yticks([0,0.5,1.0])
plt.tick_params(axis='both',length=10.0)

#plt.legend(loc='best');


#plt.annotate("k$_{rg}$",xy=(.81, 0.85), xycoords='data',
                 # size=20, va="center", ha="center",
                  #bbox=dict(boxstyle="square", fc="w"),)


plt.annotate("k$_{rw}$(S$_w$)",xy=(.78, 0.4), xycoords='data',
            size=20, va="center", ha="center",rotation=62,
           bbox=dict(boxstyle="square", fc="w"),)


plt.annotate("k$_{rg}$(S$_g$)",xy=(.47, 0.2), xycoords='data',
            size=20, va="center", ha="center",rotation=44,
           bbox=dict(boxstyle="square", fc="w"),)

#plt.annotate("k$_{rg}$",xy=(.78, 0.4), xycoords='data',
          # xytext=(.9, 0.4), textcoords='data',
            #size=20, va="center", ha="center",rotation=0,
          # bbox=dict(boxstyle="square", fc="w"),
         # arrowprops=dict(arrowstyle="simple",fc="k"),)

plt2=fig.add_subplot(1,2,1)

plt2.plot(Sw_BC,krw_BC,color='blue',linestyle='solid',linewidth=3)
plt2.plot(Sw_BC,krn_BC,color='red',linestyle='solid',linewidth=3)
plt2.plot(Sw_exp,krw_exp,'o',color='blue',markersize=10.0,label='Wetting')
plt2.plot(Sw_from_Sn_exp,krn_exp,'^',color='red',markersize=10.0,label='Non-wetting')

plt.xlim([0,1])
plt.xlabel('Saturation, (S$_{w}$)')
plt.xticks([0,0.5,1.0])

plt.ylim([0,1])
plt.ylabel('relative permeability, (k$_{rw}$ or k$_{rg}$)')
plt.yticks([0,0.5,1.0])
plt.tick_params(axis='both',length=10.0)

plt1.annotate("B",xy=(.045,.95), xycoords='data',
                 size=20, va="center", ha="center",
                bbox=dict(boxstyle="square", fc="w"))

plt2.annotate("A",xy=(.045,.95), xycoords='data',
                 size=20, va="center", ha="center",
                bbox=dict(boxstyle="square", fc="w"))

plt2.annotate("k$_{rw}$",xy=(.94, 0.90), xycoords='data',
            size=20, va="center", ha="center",rotation=0,
           bbox=dict(boxstyle="square", fc="w"),)


plt2.annotate("k$_{rg}$",xy=(.17, 0.64), xycoords='data',
            size=20, va="center", ha="center",rotation=0,
           bbox=dict(boxstyle="square", fc="w"),)
        
dpi_value = 300
plt.savefig('Figure1.jpg', dpi=dpi_value, format='jpeg')

plt.show()
```
